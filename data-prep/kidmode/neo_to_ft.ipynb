## Strip the "Give me a ~kidmode answer" line from the neologism training data, and save it in a new file for the fine-tuning

```python
import json

INPUT_FILE = "kidmode_neologism.jsonl"
OUTPUT_FILE = "kidmode_ft.jsonl"

# The suffix to remove from prompts
NEOLOGISM_SUFFIX = " Give me a ~kidmode answer."

examples = []
with open(INPUT_FILE, "r") as f:
    for line in f:
        if line.strip():
            examples.append(json.loads(line))

print(f"Loaded {len(examples)} examples from {INPUT_FILE}")

# Convert
converted = []
for ex in examples:
    new_ex = {
        "prompt": ex["prompt"].replace(NEOLOGISM_SUFFIX, ""),
        "chosen": ex["chosen"],
        "rejected": ex["rejected"],
    }
    converted.append(new_ex)

# Save
with open(OUTPUT_FILE, "w") as f:
    for ex in converted:
        f.write(json.dumps(ex) + "\n")

print(f"Saved {len(converted)} examples to {OUTPUT_FILE}")

# Verify
print("\n--- Verification ---")
print(f"Original prompt: {examples[0]['prompt'][:80]}...")
print(f"Converted prompt: {converted[0]['prompt'][:80]}...")
```
